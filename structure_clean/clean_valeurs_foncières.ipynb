```python
import pandas as pd

path = "../data/valeurs_foncieres_2022.csv"
```

# Infos

```python
df = pd.read_csv(path)
df.head(5)  #affiche 5 premiere ligne
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check différents types (choix multiples)
for col in df:
    if not (pd.api.types.is_numeric_dtype(df[col])):
        print('\n', col)
        if len(df[col].unique()) < 10:
            print(df[col].unique())
```

# Clean général

```python
# Colonnes vides 
empty_cols = df.columns[df.isna().all()].tolist()
print(empty_cols)
for col in empty_cols:
    df.drop(col, axis=1, inplace=True)
```

```python
# Colonne inutile (puisque toujours égale à 69)
df.drop('code_departement', axis=1, inplace=True)
```

```python
# Supprimer les colonnes entièrement vides
df = df.dropna(axis=1, how="all")

# Colonnes négligeables (presques vides)
for col in df:
    if df[col].isnull().sum() / df[col].shape[0] > 0.75:
        print(col)
        df.drop(col, axis=1, inplace=True)
```

```python
# Supression doublons 
df.drop_duplicates(inplace=True)
```

```python
# Harmoniser texte (en minuscule)
for col in df:
    if not (pd.api.types.is_numeric_dtype(df[col])):
        df[col] = df[col].str.lower()

# Standardiser les noms de colonnes
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# object -> datetime // float -> int 
for col in df:
    if 'date' in col.lower():
        df[col] = pd.to_datetime(df[col])
    if ('code' in col.lower()) and (pd.api.types.is_numeric_dtype(df[col])):
        df[col] = df[col].astype('Int64')
```

```python
# Valeurs manquantes 
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]

df_median = df.copy()
for col, val in missing_values.items():
    if pd.api.types.is_numeric_dtype(df[col]):
        df_median[col].fillna(df_median[col].median(), inplace=True)
    else:
        print(f"{col}: {val}")
```

# Clean spécialisé

```python
# Check valeurs min
if df['valeur_fonciere'].min() < 0:
    print(f"Valeur foncière négative : {df['valeur_fonciere'].min()}")
    for value in df['valeur_fonciere']:
        if value < 0:
            value = 0

if df['nombre_lots'].min() < 0:
    print(f"Nombre de lots négative : {df['nombre_lots'].min()}")
    for value in df['nombre_lots']:
        if value < 0:
            value = 0

if df['surface_reelle_bati'].min() < 0:
    print(f"Surface réelle bati négative : {df['surface_reelle_bati'].min()}")
    for value in df['surface_reelle_bati']:
        if value < 0:
            value = 0

if df['nombre_pieces_principales'].min() < 0:
    print(f"Nombre de pieces principales négative : {df['nombre_pieces_principales'].min()}")
    for value in df['nombre_pieces_principales']:
        if value < 0:
            value = 0

if df['surface_terrain'].min() < 0:
    print(f"Surface du terrain négative : {df['surface_terrain'].min()}")
    for value in df['surface_terrain']:
        if value < 0:
            value = 0
```

```python
df.drop('id_mutation', axis=1, inplace=True)
df.drop('date_mutation', axis=1, inplace=True)
df.drop('numero_disposition', axis=1, inplace=True)
df.drop('nature_mutation', axis=1, inplace=True)
df.drop('adresse_numero', axis=1, inplace=True)
df.drop('adresse_nom_voie', axis=1, inplace=True)
df.drop('adresse_code_voie', axis=1, inplace=True)
df.drop('id_parcelle', axis=1, inplace=True)
df.drop('lot1_numero', axis=1, inplace=True)
df.drop('nombre_lots', axis=1, inplace=True)
df.drop('code_type_local', axis=1, inplace=True)
df.drop('code_nature_culture', axis=1, inplace=True)
df.drop('nature_culture', axis=1, inplace=True)
```
